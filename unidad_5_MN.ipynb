```python
import numpy as np  # Importa NumPy para manejo de arreglos y operaciones

x = np.array([1, 2, 3])   # Define los valores de x conocidos
y = np.array([1, 4, 9])   # Define los valores de y correspondientes

n = len(x)  # Calcula el número de puntos

tabla = np.zeros((n, n))  # Inicializa la tabla de diferencias divididas

tabla[:, 0] = y  # Llena la primera columna con los valores de y

for j in range(1, n):  # Itera sobre las columnas de la tabla
    for i in range(n - j):  # Itera sobre las filas disponibles
        tabla[i, j] = (tabla[i + 1, j - 1] - tabla[i, j - 1]) / (x[i + j] - x[i])

print("Tabla de diferencias divididas:\n", tabla)  # Muestra la tabla
```

```python
import numpy as np

#Evaluacion de polinomio de newtoon

x = np.array([1,2,3]) #define los valores de x conocidos
y = np.array([1,4,9]) #define los valores de y correspondientes

n = len(x) #Calcula el numero de puntos
tabla = np.zeros((n,n)) #Inicializa la tabla de diferencias divididas


tabla[:,0] = y #Llena la primera columna con valores de y

for j in range(1,n): #Itera sobre las columnas de la tabla
  for i in range(n-j): #Itera sobre las filas disponibles
    tabla[i,j] = (tabla[i+1,j-1] - tabla[i,j-1])/(x[i+j] - x[i]) #calcula la diferencia


def newton_eval(x_data, tabla, x_val): #Define funcion para evaluar el polinomio
  n = len(x_data) #numero de puntos
  resultado = tabla[0,0] #Inicializa el primer coeficiente

  producto = 1 #Inicializa el producto acumulado

  for i in range(1,n): #Itera sobre terminos del polinomio
    producto *= (x_val - x_data[i-1]) #Calcula el termino (x-x_i)
    resultado += tabla[0,i]*producto #Suma el termino correspondiente

  return resultado #Retorna valor interpolado

valor = newton_eval(x,tabla,2.5) #Evalua el polinomio en x=2.5

print("Valor Interpolado:",valor) #Muestra resultado
```

```python
# interpolacion de datos
import numpy as np
x = np.array ([ 0 , 1 , 2 , 3 ]) # datos de entrada
y =  np.array ([1 , 2 , 0 , 5 ])# datos observados ( puedes ser incompletos o ruidoss)

n = len(x) #Calcula el numero de puntos
tabla = np.zeros((n,n)) #Inicializa la tabla de diferencias divididas

tabla[:,0] = y #Llena la primera columna con valores de y

for j in range(1,n): #Itera sobre las columnas de la tabla
  for i in range(n-j): #Itera sobre las filas disponibles
    tabla[i,j] = (tabla[i+1,j-1] - tabla[i,j-1])/(x[i+j] - x[i]) #calcula la diferencia


def newton_eval(x_data, tabla, x_val): #Define funcion para evaluar el polinomio
  n = len(x_data) #numero de puntos
  resultado = tabla[0,0] #Inicializa el primer coeficiente

  producto = 1 #Inicializa el producto acumulado

  for i in range(1,n): #Itera sobre terminos del polinomio
    producto *= (x_val - x_data[i-1]) #Calcula el termino (x-x_i)
    resultado += tabla[0,i]*producto #Suma el termino correspondiente

  return resultado #Retorna valor interpolado

x_interp = 1.5 # punto donde se quiere estimar el valor

y_interp = newton_eval (x, tabla , x_interp) #interpolacion

print("valor estimado:", y_interp ) # resultado final
```

```python
import numpy as np # importa numpy para el manejo de datos numericos
from scipy.interpolate import CubicSpline # importa el cubuculo

x = np.array([0,1,2,3]) # define los puntos de el eje x
y = np.array([1,3,2,5])# define los valores correspondientes

spline = CubicSpline(x,y) # construye el spline cubico

x_interp = np.linspace(0,3,100) # genera 100 puntos para evaluar

y_interp = spline(x_interp) # evalua el spline en los puntos

print ("valores importados:", y_interp) # muestra valores
```

```python
import numpy as np # importa numpy

x = np.array([1,2,3,4,5,]) # define los puntos de el eje x
y = np.array([2,4,5,4,5])# define los valores correspondientes

coef = np.polyfit ( x,y,1) # ajusta un polinomio de grado

m = coef[0] # pendiente de la recta
b = coef [1] # interceccion en el eje y

y_pred = m*x + b # calcula los valores predichos por el momento

print ("pendiente:",m)# muestra la pendiente
print ("intercepto:",b ) # muestra el termino independiente
print ( "prediccion:",y_pred) # muestra los valores estimados
```
