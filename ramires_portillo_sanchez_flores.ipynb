```python
# Paso 1) Diseñamos una señal sintética (suma de senoidales + ruido)

import numpy as np
import matplotlib.pyplot as plt

# Frecuencia de muestreo y duración
fs = 500        # Hz
T = 2.0         # segundos

# Vector de tiempo
t = np.linspace(0, T, int(fs*T), endpoint=False)

# Frecuencias de las señales
f1, f2 = 50, 120  # Hz

# Señal sintética: combinación de dos senoidales + ruido gaussiano
signal = 1.0 * np.sin(2*np.pi*f1*t) + \
         0.6 * np.sin(2*np.pi*f2*t) + \
         0.3 * np.random.normal(0, 1, len(t))

# Paso 2) FFT (solo frecuencias positivas con rfft)
Y = np.fft.rfft(signal)
freqs = np.fft.rfftfreq(len(t), d=1/fs)
amplitud = np.abs(Y) / len(t)

# Paso 3) Espectro de amplitud
fig = plt.figure(figsize=(8, 4))
ax = plt.gca()
ax.plot(freqs, amplitud, color='purple', linewidth=2)

# Enfocamos la ventana de interés
ax.set_xlim(0, 200)

# Etiquetas y título
ax.set_title('🔍 Análisis Espectral: Picos de Frecuencia')
ax.set_xlabel('Frecuencia (Hz)')
ax.set_ylabel('Amplitud')

# Cuadrícula y estilo
ax.grid(True, linestyle='--', alpha=0.5)

# Líneas de referencia en las frecuencias principales
ax.axvline(f1, color='red', linestyle='--', alpha=0.7, label=f'{f1} Hz')
ax.axvline(f2, color='orange', linestyle='--', alpha=0.7, label=f'{f2} Hz')
ax.legend()

plt.show()
```
