```python
# metodo de aproximaciones sucesivas
import numpy as np # biblioteca para calculos numericos

def g(x): # definicion de la funcion iterativa  g(x)
    return np.cos(x) #funcion de coseno que define el punto fijo

x = 0.5 # valor inicial de la aproximacion
tolerancia = 1e-6 # error maximo permitido
max_iter = 100 # numero maximo de iteraciones

for i in range(max_iter):# ciclo de iteraciones
    x_new = g(x) # aplicacion de la formula de aproximacion susesiva

    if abs(x_new - x) < tolerancia: # criterio de paro basado en el erro absoluto
        break # termina si la diferencia es suficientemente pequeña

    x = x_new # actualizada del valor para la siguiennte iteracion

print("raiz aproximada :", x_new)# mostrar resultado final
print("iteracionees realizadas :", i + 1)#mostrar numero de iteraciones
```

```python
# inteporlacion manual lineal
# datos (1,2),(3,6)
#interpolar en x=2

x0 = 1 #pimer punto en x
y0 = 2 #valor correspondiente en y
x1 = 3 #segundo punto en x
y1 = 6 #valor correspondiente en y

x = 2 #punto donde se desea interpolar

p = y0 + ((y1 - y0) / (x1 - x0)) * (x - x0) # formula de interpolacion linwal

print (" valor interpolado:",p) # mostrar resultado
```

```python
#Interpolacion con Numpy
import numpy as np #Biblioteca numerica

x_datos = np.array([0,1,2]) # arreglo de valores conocidos en x
y_datos = np.array([1,3,2])# areglo de valores conocidos en y

coeficientes = np.polyfit(x_datos, y_datos, 2)# ajuste de polinomio
polinomio = np.poly1d(coeficientes)#construccion de polinomio

x_eval = 1.5 # punto donde se evaluara el polinomio
y_eval = polinomio(x_eval) # evaluacion del polinomio

print("Coeficiente del polinomio", coeficientes)# mostrar coeficientes
print("valor interpolado :",y_eval)# mostrar valor interpolado
```

```python
#IMPLEMENTAR BISECCION MANUAL
import numpy as np
def f(x):  #definicion de la funcion a analizar
    return x**3 - x - 2 #expresion algebraica de la funcion

a = 1   #extremo izquierdo inicial
b = 2   #extremo derecho inicial
tolerancia = 1e-6   #error maxima permitido
max_iter = 23   #numero maximo de iteraciones permitidas

for i in range(max_iter):   #ciclo de iteraciones
    c = (a + b) / 2   #calculo del punto medio
    fc = f(c)   #evaluacion de la funcion en el punto medio

    if abs(fc) < tolerancia:   #criterio de paro por valor cercano a cero
        break   #termina el ciclo

    if f(a) * fc < 0:   #verifica si la raiz esta en el subintervalo
        b = c   #actualiza extremo derecho
    else:
        a = c   #actualiza el extremo izquierdo

print("raiz aproximada: ",c)   #muestra la raiz
print("iteraciones realizadas: ", i+1)   #numero de iteraciones
```
