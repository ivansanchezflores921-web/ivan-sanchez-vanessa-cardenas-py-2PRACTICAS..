```python
#ejercicio: aproximacion numerica de una raiz
import numpy as pd

x = pd.linspace(0,2,1000) # se crea un arreglo de 1000 valores
f = x** 2 - 2 # se define la funcion

x_aprox = x[pd.argmin(pd.abs(f))] # se toma el valor de x
print(x_aprox)# se imprime la aproximacion
```

```python
import numpy as np# se importa las librerias

mediciones = np.array([9.8 , 9.81 , 9.79 , 9.80 ]) # se crea un arreglo con las mediciones
valor_real = 9.81 # se calcula el valor real

media=np.mean(mediciones) # se calcula la media
error = abs(media-valor_real)# se calcula el error absoluto
# se imprimen los do resultados
print("media;",media)
print("error absoluto:",error)
```

```python
import math# se importa el modulo math

x = 1 # se define el valor de x
aprox = 1 + x + (x**2)/2# se calcula la aproximacion
real = math.exp(x)  # Definimos el valor real (ej. e^x)
error = abs(real - aprox)# se calcula el error absoluto
# se imprimen los resultados
print("aprox:", aprox)
print("real:", real)
print("error:", error)
```

```python
x = 2# se le da valor a x
dx = 0.01# se define el error
# se calcula el valor de la funcion
y = x**2
# El error propatado dy es aproximadamente |f'(x)| * dx
dy = abs(2 * x) * dx
# se imprimen los resultados
print("y:", y)
print("error aproximado en y (dy):", dy)
```

```python
import numpy as np# se importa las librerias
# Definimos la matriz A con corchetes anidados para 2D
A = np.array([[3, 2], # se definen las matrices
              [1, 2]])

b = np.array([5, 5])# definimos el vector

# Resolvemos el sistema de ecuaciones Ax = b
x = np.linalg.solve(A, b)
```

```python
from scipy.optimize import minimize#se importa la funcion que minimize scipy.optimize

f = lambda x: (x - 2)**2# se define la funcion para minimizar

resultado = minimize(f, x0=0)# se aplica el metodo de minimizacion
# se imprime
print("Solución (x):", resultado.x)

# Corregido: se añade el paréntesis de cierre
print(resultado)
```
