```python
import numpy as np

# Define la ecuación diferencial
# dy/dx = x + y

def f(x, y):
    return x + y

# Valores iniciales
x = 0
y = 1
h = 0.1

# Iteraciones
for i in range(5):
    y = y + h * f(x, y)
    x = x + h

    print("x =", x, " y =", y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define la ecuación diferencial

def f(x, y):
    return x + y

# Valores iniciales
x = 0
y = 1
h = 0.1

# Listas para almacenar resultados
x_vals = [x]
y_vals = [y]

# Método de Euler
for i in range(10):
    y = y + h * f(x, y)
    x = x + h

    x_vals.append(x)
    y_vals.append(y)

# Graficar resultados
plt.plot(x_vals, y_vals, marker='o')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solución aproximada de la EDO')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Modelo simple de aprendizaje
# dy/dt = -0.5y

def f(t, y):
    return -0.5 * y

# Valores iniciales
h = 0.1
t = 0
y = 1

# Almacenar resultados
T = [t]
Y = [y]

# Método de Euler
for i in range(20):
    y = y + h * f(t, y)
    t = t + h

    T.append(t)
    Y.append(y)

# Gráfica
plt.plot(T, Y)
plt.xlabel('Tiempo')
plt.ylabel('Aprendizaje')
plt.title('Dinámica de aprendizaje')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ecuación diferencial

def f(x, y):
    return y

# Solución exacta

def exacta(x):
    return np.exp(x)

# Valores iniciales
x = 0
y = 1
h = 0.1

x_vals = []
y_euler = []
y_exacta = []

# Método de Euler
for i in range(10):
    y = y + h * f(x, y)
    x = x + h

    x_vals.append(x)
    y_euler.append(y)
    y_exacta.append(exacta(x))

# Graficar comparación
plt.plot(x_vals, y_euler, label='Euler')
plt.plot(x_vals, y_exacta, label='Exacta')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Definición de la EDO

def f(x, y):
    return x + y

# Valores iniciales
x = 0
y = 1
h = 0.1

# Método RK2
for i in range(10):
    k1 = f(x, y)
    k2 = f(x + h, y + h * k1)

    y = y + (h / 2) * (k1 + k2)
    x = x + h

    print("x =", x, " y =", y)
```

```python
import numpy as np

# EDO

def f(x, y):
    return x + y

# Valores iniciales
x = 0
y = 1
h = 0.1

# Método RK4
for i in range(10):
    k1 = f(x, y)
    k2 = f(x + h/2, y + h*k1/2)
    k3 = f(x + h/2, y + h*k2/2)
    k4 = f(x + h, y + h*k3)

    y = y + (h/6)*(k1 + 2*k2 + 2*k3 + k4)
    x = x + h

    print("x =", x, " y =", y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# EDO

def f(x, y):
    return y

# Valores iniciales
x0 = 0
y0 = 1
h = 0.1

# Euler
x = x0
y = y0

x_euler = [x]
y_euler = [y]

for i in range(10):
    y = y + h * f(x, y)
    x = x + h

    x_euler.append(x)
    y_euler.append(y)

# RK4
x = x0
y = y0

x_rk4 = [x]
y_rk4 = [y]

for i in range(10):
    k1 = f(x, y)
    k2 = f(x + h/2, y + h*k1/2)
    k3 = f(x + h/2, y + h*k2/2)
    k4 = f(x + h, y + h*k3)

    y = y + (h/6)*(k1 + 2*k2 + 2*k3 + k4)
    x = x + h

    x_rk4.append(x)
    y_rk4.append(y)

# Graficar
plt.plot(x_euler, y_euler, label='Euler')
plt.plot(x_rk4, y_rk4, label='RK4')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Modelo simple de actualización de pesos

def f(t, w):
    return -0.3 * w

# Valores iniciales
w = 5
t = 0
h = 0.1

T = [t]
W = [w]

# Método RK4
for i in range(20):
    k1 = f(t, w)
    k2 = f(t + h/2, w + h*k1/2)
    k3 = f(t + h/2, w + h*k2/2)
    k4 = f(t + h, w + h*k3)

    w = w + (h/6)*(k1 + 2*k2 + 2*k3 + k4)
    t = t + h

    T.append(t)
    W.append(w)

# Graficar
plt.plot(T, W)
plt.title('Dinámica de pesos')
plt.xlabel('Tiempo')
plt.ylabel('Peso')
plt.grid(True)
plt.show()
```
